```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
#Read the dataset into the Pandas DataFrame!
#Does the dataset include any missing values? If so, delete the missing values entries!
breast_cancer = pd.read_csv('data.csv')
breast_cancer.dropna(axis='columns', inplace=True)
breast_cancer.head()
```

```python
#Dropping Unnecessary Columns
breast_cancer.drop(columns=['id'], inplace= True)
```

```python
breast_cancer.dtypes
```

```python
#preprocessing for the first column diagnosis
#because this is output, I will use LabelEncoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
breast_cancer['diagnosis'] = encoder.fit_transform(breast_cancer['diagnosis'])
```

```python
breast_cancer['diagnosis'].value_counts()
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
breast_cancer.loc[:, breast_cancer.columns != 'diagnosis'] = min_max_scaler.fit_transform(breast_cancer.loc[:, breast_cancer.columns != 'diagnosis'])

```

```python
sns.violinplot(x='diagnosis', y='radius_mean', data= breast_cancer)
```
