```python
import pandas as pd
```

```python
#Read the dataset into the Pandas DataFrame!
football = pd.read_csv("results.csv")
```

```python
#Does the dataset include any missing values? If so, delete the missing values entries!
football=football.dropna()
```

```python
#How many tuples are there in the dataset?
len(football)
```

```python
#How many tournaments are there in the dataset?
len(football['tournament'].unique())
```

```python
#Convert the column date to timestamps
pd.to_datetime(football['date'])
#Find out how many matches in the dataset were played in 2018.
#pd.DatetimeIndex(football['date']).year
sum(pd.to_datetime(football['date']).dt.year == 2018)
```

```python
#Calculate how many times the home team won, lost or had a draw.
import numpy as np
conditions = [football['home_score'] > football['away_score'], football['home_score'] == football['away_score'], football['home_score'] < football['away_score']]
choices = ['Won', 'draw', 'lost']
football['compare'] = np.select(conditions, choices)
```

```python
football['compare'].value_counts()
```

```python
#Plot the numbers extracted from step 5 in a pie chart.
football['converted_date']=pd.to_datetime(football['date']).dt.year
```

```python
football.converted_date.value_counts().plot(kind='pie')
```
